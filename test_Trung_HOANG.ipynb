2 Data Exploration

In the file, we observed that there are some columns that just have only two values: 
    'CAR_USE' has only 'Private' and 'Commercial'
    'MSTATUS' has only 'z_No' and 'Yes'
    'PARENT1' has only 'No' and 'Yes'
    'RED_CAR' has only 'no' and 'yes'
    'REVOKED' has only 'No' and 'Yes'
    'SEX' has only 'z_F' and 'M'
    'URBANICITY' has only 'Highly Urban/ Urban' and 'z_Highly Rural/ Rural'
    
    Therefore, we will transform these variable to binary form of 0 and 1. We do for both of train_auto and test_auto files.

```python
df_train["CAR_USE"] = df_train["CAR_USE"].replace(to_replace =["Private","Commercial"],
                 value =["0","1"]).astype(float, errors = 'raise')
df_test["CAR_USE"] = df_test["CAR_USE"].replace(to_replace =["Private","Commercial"],
                 value =["0","1"]).astype(float, errors = 'raise')
```

```python
df_train["MSTATUS"] = df_train["MSTATUS"].replace(to_replace =["z_No","Yes"],
                 value =["0","1"]).astype(float, errors = 'raise')
df_test["MSTATUS"] = df_test["MSTATUS"].replace(to_replace =["z_No","Yes"],
                 value =["0","1"]).astype(float, errors = 'raise')
```

```python
df_train["PARENT1"] = df_train["PARENT1"].replace(to_replace =["No","Yes"],
                 value =["0","1"]).astype(float, errors = 'raise')
df_test["PARENT1"] = df_test["PARENT1"].replace(to_replace =["No","Yes"],
                 value =["0","1"]).astype(float, errors = 'raise')
```

```python
df_train["RED_CAR"] = df_train["RED_CAR"].replace(to_replace =["no","yes"],
                 value =["0","1"]).astype(float, errors = 'raise')
df_test["RED_CAR"] = df_test["RED_CAR"].replace(to_replace =["no","yes"],
                 value =["0","1"]).astype(float, errors = 'raise')
```

```python
df_train["REVOKED"] = df_train["REVOKED"].replace(to_replace =["No","Yes"],
                 value =["0","1"]).astype(float, errors = 'raise')
df_test["REVOKED"] = df_test["REVOKED"].replace(to_replace =["No","Yes"],
                 value =["0","1"]).astype(float, errors = 'raise')
```

```python
df_train["SEX"] = df_train["SEX"].replace(to_replace =["z_F","M"],
                 value =["0","1"]).astype(float, errors = 'raise')
df_test["SEX"] = df_test["SEX"].replace(to_replace =["z_F","M"],
                 value =["0","1"]).astype(float, errors = 'raise')
```

```python
df_train["URBANICITY"] = df_train["URBANICITY"].replace(to_replace =["Highly Urban/ Urban","z_Highly Rural/ Rural"],
                 value =["0","1"]).astype(float, errors = 'raise')
df_test["URBANICITY"] = df_test["URBANICITY"].replace(to_replace =["Highly Urban/ Urban","z_Highly Rural/ Rural"],
                 value =["0","1"]).astype(float, errors = 'raise')
```

Also, there are the columns that their values should delete the special string '$' and ',', then we could convert the data type. We apply for both of train_auto and test_auto files.

```python
cols_to_check = ["INCOME","HOME_VAL","BLUEBOOK","OLDCLAIM"]
for col in cols_to_check:
    df_train[col] = df_train[col].map(lambda x: str(x).lstrip('$'))
    df_train[col] = df_train[col].str.replace(',', '').astype(float)
    df_test[col] = df_test[col].map(lambda x: str(x).lstrip('$'))
    df_test[col] = df_test[col].str.replace(',', '').astype(float)
```

3 Data mining and imputation

We check the correlation betweens the variables. We plot also the correlation matrix:

```python
import seaborn as sn
```

```python
df_train.corr(method ='pearson')
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(20,20))
sn.heatmap(df_train.corr(method ='pearson'), annot=True,linewidths=.5, ax=ax)
```

From the correlation result, we look for the variable that may have strong affect to the TARGET_FLAG. We will sort them in order: 

```python
df_train.corr(method ='pearson').TARGET_FLAG.abs().sort_values()
```

```python
df_train.corr(method ='pearson').TARGET_FLAG.sort_values()
```

From the absolute ranking result, we may choose several variables that correlate to TARGET_FLAG such as 'URBANICITY','MVR_PTS','CLM_FREQ','HOME_VAL', 'PARENT1','REVOKED' and 'CAR_USE'. Noting that we do not use TARGET_AMT since it is null in the test_auto file. 

```python
target_variables = ['URBANICITY','MVR_PTS','CLM_FREQ','HOME_VAL', 'PARENT1','REVOKED','CAR_USE']
```

```python
df_train[target_variables].isnull().sum()
```

```python
df_test[target_variables].isnull().sum()
```

We observe that the variable HOME_VAL has the missing values in both files. Then it lead us to remove NaN values and impute the missing values. One suggestion is to use imputation by mean values.

```python
from sklearn.preprocessing import Imputer
imp=Imputer(missing_values="NaN", strategy="mean" )
df_train["HOME_VAL"]=imp.fit_transform(df_train[["HOME_VAL"]]).ravel()
df_test["HOME_VAL"]=imp.fit_transform(df_test[["HOME_VAL"]]).ravel()
```

```python
df_train["HOME_VAL"]
```

4 Builing the logistic regression

We observe that our problem is supervised learning. Then we will build logistic regression that could predict 'TARGET_FLAG'

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

In the train_auto file, the predicted colums is TARGET_FLAG while the variable colums are chosen due to the correlation's affection.

```python
#X = df_train[['URBANICITY','MVR_PTS','CLM_FREQ','HOME_VAL', 'PARENT1']]
X = df_train[target_variables]
y = df_train['TARGET_FLAG']
```

We standardized the data first to get better prediction: 

```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
X = pd.DataFrame(scaler.transform(X),columns=X.columns)
```

```python
X
```

Using the train_test_split from sklearn.model_selection. We set the test size to 0.2, and it means that the model testing will be based on 20% of the dataset, while the model training will be based on 80% of the dataset:

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

Apply the logistic regression as follows:

```python
logistic_regression= LogisticRegression()
logistic_regression.fit(X_train,y_train)
```

Taking the prediction from the test data

```python
y_pred=logistic_regression.predict(X_test)
```

```python
y_pred
```

To evaluate the model performance, we print out the accuracy and plot the confusion matrix as the appropriate metric for our problem:

```python
print('Model Accuracy:',metrics.accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
import seaborn as sns
cm=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(12,6))
plt.title("Confusion Matrix")
sns.heatmap(cm, annot=True,fmt='d', cmap='Blues')
plt.ylabel("Actual Values")
plt.xlabel("Predicted Values")
plt.savefig('confusion_matrix.png')
```
